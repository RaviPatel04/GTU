```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5,6]
y = [4,6,5,3,1,2]
color = [20,30,40,50,60,70]
plt.plot(x,y,color = "olive") 
plt.title("line Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No.",fontsize = 15)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [4,6,5,3,1,2]
color = [20,30,40,50,60,70]
plt.scatter(x,y,c = color,cmap = "hsv") 
plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No.",fontsize = 15)
plt.show()
```

```python
x = ["Python","Java","c","c++","Mojo"]
y = [4.8,4.4,3.7,3.2,2]
plt.xlabel("Language",fontsize = 12)
plt.ylabel("Rating",fontsize = 12)
plt.title("Most like lang.")
plt.bar(x,y,width = .5,color = ["r","b","g","y","c"],edgecolor = "k",linewidth = 2)
plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x,labels = ["1st"],patch_artist = True)
plt.show()
```

```python
x = np.random.randint(10,80,(50))
plt.hist(x,color = "coral",edgecolor = "k")
plt.title("histogram plot")
plt.xlabel("number")
plt.ylabel("repetition")
plt.show()
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]
plt.pie(x,labels = y,explode = [0.2,0,0,0],colors = ["b","lime","g","peru"],autopct = "%0.1f%%",labeldistance = 1.12)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt  


ug = nx.Graph()
ug.add_nodes_from([1,2,3,4])
ug.add_edges_from([(1,2),(2,3),(3,1),(1,4)])


dg = nx.DiGraph()
dg.add_nodes_from(['A','B','C','D'])
dg.add_edges_from([('B', 'A'),('B', 'C'),('C', 'A'),('A','D')])


nx.draw(ug, with_labels=True, node_color='gold', node_size=800)
plt.title("Undirected Graph")
plt.show()

nx.draw(dg, with_labels=True, node_color='silver', node_size=800)
plt.title("Directed Graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3,4,5,6,7,8]
y = [0,1,1.5,0.5,1,-1.8,-1,1,1.9]
plt.plot(x,y)
plt.xlim(-2,12)
plt.ylim(-2,2)
plt.title("Tata Share",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("Price",fontsize = 15)
plt.text(8,1.8,"High",fontsize = 11,color = "b")
plt.annotate("Low",xy = (5,-1.8),xytext = (8,-1.5), arrowprops = dict(facecolor = "r",shrink = 10))
plt.show()
```

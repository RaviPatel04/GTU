```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [2,1,4,3]

plt.figure(dpi = 300)
plt.subplot(2,3,1)
plt.plot(x,y,color = "r")

plt.subplot(2,3,2)
plt.scatter(x,y)

plt.subplot(2,3,3)
plt.bar(x,y)

plt.subplot(2,3,4)
plt.boxplot(x)

plt.subplot(2,3,5)
plt.hist(x)

plt.subplot(2,3,6)
plt.pie(x,autopct = "%0.0f%%")

plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt  


ug = nx.Graph()
ug.add_nodes_from([1,2,3,4,5,6,7,8])
ug.add_edges_from([(1,2),(1,3),(1,4),(2,3),(2,5),(2,6),(3,6),(5,6),(4,7),(4,8),(7,8)])


dg = nx.DiGraph()
dg.add_nodes_from(['A','B','C','D','E','F'])
dg.add_edges_from([('A', 'B'),('A', 'D'),('B', 'C'),('C','D'),('C','E'),('D','E')])


nx.draw(ug, with_labels=True, node_color='gold', node_size=800)
plt.title("Undirected Graph")
plt.show()

nx.draw(dg, with_labels=True, node_color='silver', node_size=800)
plt.title("Directed Graph")
plt.show()
```

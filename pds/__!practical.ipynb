```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import random
```

```python
x = [1,2,3,4,5,6,7]
y = [1.5,1,2,4.5,.5,2,5]
# plt.subplots(dpi = 300)
plt.figure(dpi = 300,facecolor = "gold")
plt.plot(x,y,linestyle = "-.",color = "purple",marker = "d",markeredgecolor = "k",label = "s_p")
plt.text(4.1,4.5,"high",color = "r",bbox = dict(facecolor = "y",alpha = .5,edgecolor= "b"))
plt.annotate("low",xy = (5,.5),xytext = (6.5,1),arrowprops = dict(facecolor = "r"))
plt.xlabel("X-axis",fontsize = 12)
plt.ylabel("Y-axis",fontsize = 12)
plt.title("Line-plot",fontsize = 12)
plt.legend(facecolor = "aqua",edgecolor = "maroon",shadow = True)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,3,1,6,4]
plt.figure(dpi = 300,facecolor = "silver")
plt.scatter(x,y,c= [20,40,60,80,100],s = [30,35,40,45,50],cmap = "hsv",edgecolor = "k",label = "point")
plt.colorbar(label = "Colorbar")
plt.title("Scatter-plot")
plt.xlabel("X-axis")
plt.ylabel("Yiaxis")
plt.legend(loc = 2,shadow = True,facecolor = "aqua",edgecolor = "maroon")
plt.show()
```

```python
x = ["c","c++","java","python"]
y = [10,20,30,40]
plt.figure(dpi = 300,facecolor = "skyblue")
plt.bar(x,y,width = .5,color = ["r","b","g","y"],edgecolor = "indigo",linewidth= 2,align = "edge",label = "Most like lang",linestyle = ":",alpha = .5)
plt.title("Bar-plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(shadow = True, facecolor = "aqua", edgecolor = "maroon")
plt.show()
```

```python
x = np.random.randint(10,50,(50))
plt.figure(dpi = 300,facecolor = "bisque")
plt.hist(x,edgecolor = "k",bins = [10,15,20,25,30,35,40,45,50],label = "x-data",color = "r")
plt.title("Histogram-plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(shadow = True,facecolor = "aqua",edgecolor = "maroon")
# plt.grid()
plt.show()
```

```python
x = [10,20,30,40]
y = ["c","c++","java","python"]
plt.figure(dpi = 300,facecolor = "palegreen")
plt.title("Pie-chart")
plt.pie(x,labels = y,colors = ["r","b","y","g"],explode = [.2,0,0,0],autopct = "%.1f%%",textprops = dict(fontsize = 15),wedgeprops = dict(edgecolor = "k",linewidth = 2), shadow = True)
plt.show()
```

```python
x = [10,20,30,40,50,100]
plt.figure(dpi = 300, facecolor = "powderblue")
plt.title("Box-plot")
plt.boxplot(x,patch_artist = True,sym = "r",notch = True,boxprops = dict(color = "navy"),capprops = dict(color = "r"),whiskerprops = dict(color = "k"))
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

ug = nx.Graph()
ug.add_nodes_from([1,2,3,4,5])
ug.add_edges_from([(1,2),(1,4),(1,5),(2,3),(2,4),(3,4),(4,5)])
nx.draw(ug,with_labels = True,node_color = "gold",node_size = 500)
plt.title("Undirected graph")
plt.show()
A = nx.adjacency_matrix(ug).toarray()
print(A)

dg = nx.DiGraph()
dg.add_nodes_from(["a","b","c","d","e"])
dg.add_edges_from([("a","b"),("b","a"),("a","c"),("c","d"),("d","e"),("e","c")])
nx.draw(dg,with_labels = True,node_color = "silver",node_size = 500)
plt.title("Directed graph")
plt.show()
B = nx.adjacency_matrix(dg).toarray()
print(B)
```

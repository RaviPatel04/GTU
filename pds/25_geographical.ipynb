**Write a python program for plotting of geographical data using
basemap.**

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Create a Basemap object for the world map
m = Basemap(projection='merc', resolution='l',
            llcrnrlat=-60, urcrnrlat=85,
            llcrnrlon=-180, urcrnrlon=180)

fig, ax = plt.subplots(figsize=(10, 8))

m.drawcoastlines()
m.drawcountries()
m.drawstates()
m.drawparallels(range(-90, 91, 30), labels=[1, 0, 0, 0])
m.drawmeridians(range(-180, 181, 60), labels=[0, 0, 0, 1])

data_points = {
    'New York City': (40.7128, -74.0060),
    'Los Angeles': (34.0522, -118.2437),
    'London': (51.5074, -0.1278),
    'Tokyo': (35.682839, 139.759455)
}

for city, (lat, lon) in data_points.items():
    x, y = m(lon, lat)
    m.plot(x, y, 'ro', markersize=6)
    plt.text(x, y, city, fontsize=12, color='blue')

plt.title('Geographical Data Plot with Basemap')
plt.show()
```

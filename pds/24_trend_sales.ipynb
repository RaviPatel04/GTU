**Write Python program to plot time series with trends of sales.**

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('sales.csv', index_col='Date')

plt.plot(df.index, df['Sales'])
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt

sem = [5, 6, 7, 8]
stu1 = [9, 8, 9.45, 9.96]
stu2 = [8.32, 9.70, 7.50, 8]


plt.plot(sem, stu1, label='Student 1', marker='d',linestyle = "dashdot",color = "purple",markeredgecolor = "k")
plt.plot(sem, stu2, label='Student 2', marker='8',linestyle = "dashed",color = "olive",markeredgecolor = "k")

plt.xlabel('Semester')
plt.ylabel('SPI')
plt.title('Student SPI Over Semesters')

plt.legend(facecolor = "aqua",shadow = True,edgecolor = "maroon")
plt.show()
```

```python
%%timeit
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

factorial(10)
```

**With example, explain the concept of bags of words model.**

```python
from sklearn.feature_extraction.text import CountVectorizer

# Create a corpus of documents
documents = [
    "This is a document about cats.",
    "This is a document about dogs.",
    "This is a document about cats and dogs."
]

# Create a CountVectorizer object
vectorizer = CountVectorizer()

# Transform the documents into a matrix of word counts
X = vectorizer.fit_transform(documents)

# Print the vocabulary
print(vectorizer.get_feature_names_out())

# Print the document-term matrix
print(X.toarray())
```

**Explain with example how to parse XML and HTML.**

```python
import xml.etree.ElementTree as ET

# Parse XML file from disk
xml_file = 'books.xml'
tree = ET.parse(xml_file)
root = tree.getroot()

# Iterate through nested elements
for book in root.findall('book'):
    title = book.find('title').text
    author = book.find('author').text
    print(f"Book: {title}, Author: {author}")
```

**Write a python program to demonstrate the concept of skewness and kurtosis.**

```python
import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt

# Generate a sample data set (you can replace this with your own data)
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate skewness and kurtosis
skewness = skew(data)
kurt = kurtosis(data)

# Display the histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.7, color='red')
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Print skewness and kurtosis
plt.text(1.2,0.4,f'Skewness: {skewness:.2f}\nKurtosis: {kurt:.2f}', fontsize=12, bbox=dict(facecolor='white', alpha=0.7))

plt.show()
```

**Write a python code to access data from the web.**


```python
# Import the requests library
import requests

# Define the URL of the web page to access
url = "https://jsonplaceholder.typicode.com/todos/1"

# Send a GET request to the URL and store the response object
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Print the content of the web page
    print(response.text)
else:
    # Print an error message
    print("Something went wrong. The request failed with status code:", response.status_code)
```

**Write a small code to perform following operations on data: Slicing, Dicing, Concatenation, Transformation.**

```python
import pandas as pd

# Create a dataframe
df = pd.DataFrame({'A': [1, 2, 3,4], 'B': [5, 6, 7,8]})

# Slice the dataframe
df_sliced = df[:2]

# Dice the dataframe
df_diced = df[df['A'] > 2]

# Concatenate the dataframes
df_concatenated = pd.concat([df_sliced, df_diced], ignore_index=True)

# Transform the dataframe
df_transformed = df.apply(lambda x: x * 2)

# Print the results
print(f"Slicing:\n{df_sliced}")
print(f"Dicing:\n{df_diced}")
print(f"concatenate:\n{df_concatenated}")
print(f"transformed:\n{df_transformed}")
```

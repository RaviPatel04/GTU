**Draw a graph which satisfied following condition.**

** x = range(10)**

** y = cos(x)**

** graph should be “dot-dashed”**

** Give the label to x-axis “this is X dimension”**

** Give the label to y-axis “this is Y dimension”**

** Give the Title to graph “This is My First Graph“**

```python
import matplotlib.pyplot as plt
from math import cos
```

```python
x = list(range(10))
y = [cos(i) for i in x]
plt.plot(x,y,linestyle = "-.",color = "r")
plt.title("This is My First Graph")
plt.xlabel("This is X dimension")
plt.ylabel("This is Y dimension")
plt.show()
```
